```python
# Countries with maximum players in EPL
nat_bar= epl_df.groupby('Nationality').size().sort_values(ascending =False).head(10)
nat_bar.plot(kind='bar',figsize=(12,4),color = sns.color_palette('magma'))
```

```python
epl_df.head(5)
```

```python
#clubs with maximum number of players out of England

non_eng = epl_df[epl_df.Nationality!='ENG'].groupby('Club').size().sort_values(ascending=False).head(8)
non_eng.plot(kind='bar',figsize=(12,6),color = sns.color_palette('Paired'))
```

```python
# players based on age groups

age_under20 = epl_df[epl_df['Age']<=20]
age20_25 = epl_df[(epl_df['Age']>20) & (epl_df['Age']<=25)]
age25_30 = epl_df[(epl_df['Age']>25) & (epl_df['Age']<=30)]
age_above30 = epl_df[epl_df['Age']>30]
```

```python
x = np.array([age_under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),age_above30['Name'].count()])
mylabels = ['Less than 20','Between 20 and 25','Between 25 and 30','More than 30']
color = sns.color_palette('Set2')
plt.title('Players and age groups', fontsize=25)
plt.pie(x,labels=mylabels,colors=color,autopct="%.0f%%")
plt.show()
```

```python
# total young players in each clubs (under 20)
t_young_clb=age_under20.groupby('Club').size().sort_values(ascending=False).head(10)
t_young_clb.plot(kind='bar',figsize=(12,6),color = sns.color_palette('mako'))
```

```python
# clubs with older players (above 30)
t_old_clb=age_above30.groupby('Club').size().sort_values(ascending=False).head(10)
t_old_clb.plot(kind='bar',figsize=(12,6),color = sns.color_palette('rocket'))
```

```python
# Top goal scorers
epl_df.sort_values('Goals', ascending=False).head(10)

```

```python
# Top players with most assists.
top_10_assist=epl_df[['Name','Club','Matches','Assists']].nlargest(n=10, columns = 'Assists')
top_10_assist
```

```python
# Team with most number of goals
Team_goals=epl_df.groupby('Club')['Goals'].sum().sort_values(ascending=False)
Team_goals.plot(kind='bar',figsize=(14,6),color = sns.color_palette('viridis'))
```

```python
# Top 5 teams with most yellow cards
yellow_cards = epl_df.groupby('Club')['Yellow_Cards'].sum().sort_values(ascending=False).head(5)
yellow_cards.plot(kind='bar',figsize=(14,6),color = 'Yellow')
```

```python
# Top 5 teams with most red cards
yellow_cards = epl_df.groupby('Club')['Red_Cards'].sum().sort_values(ascending=False).head(5)
yellow_cards.plot(kind='bar',figsize=(14,6),color = sns.color_palette('rocket'))
```

```python
# Goals with and without assists

tot_gl_with=epl_df['Goals'].sum()
t_a_g=epl_df['Assists'].sum()
t_na_g=tot_gl_with-t_a_g
x=(t_a_g,t_na_g)
mylabels = ['Assisted goals','Non assisted goals']
color = sns.color_palette('Spectral')
plt.title('Goals with and without assists', fontsize=25)
plt.pie(x,labels=mylabels,colors=color,autopct="%.0f%%")
plt.show()
```
